## MACHINE LEARNING

```python
# Display all numeric features
movies_data_rating.select_dtypes(include = "number")
```

```python
# selects the feature columns and converts the target into a binary label for classification.
def select_columns(df: pd.DataFrame, features : list, target : list = "rating"):
  X = df[features]
  y = (df[target] >= 3).astype("int")
  return X, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = ["budget", "popularity", "vote_average"]
X, y = select_columns(movies_data_rating, features = features)
```

```python
# Split the dataset into training (70%) and testing (30%) sets to evaluate model performance consistently
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
# Fit the scaler by learning the minimum and maximum values from the training data
minmax.fit(X_train)
```

```python
X_train
```

```python
X_test
```

```python
def scaling(df, scaler):
  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array, columns = scaler.get_feature_names_out())
  return scaled_df
```

```python
X_train_scaled = scaling(X_train, minmax)
X_test_scaled = scaling(X_test, minmax)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
X_test_scaled.iloc[0]
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Create a KNN classifier using 5 nearest neighbors
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
y_train_pred_knn = knn.predict(X_train_scaled)
```

```python
y_train_pred_knn
```

```python
y_test_pred_knn = knn.predict(X_test_scaled)
```

```python
y_test_pred_knn
```

```python
pd.crosstab(y_test, y_test_pred_knn)
```

The confusion matrix above shows how well the KNN model performs in predicting the two classes. The model correctly classified 22 samples as class 0 and 584 samples as class 1. However, it also made several mistakes, 175 samples that actually belong to class 0 were incorrectly predicted as class 1, and 69 samples from class 1 were incorrectly predicted as class 0. This means the model is much better at identifying class 1 than class 0. The large number of false positives for class 0 suggests that the model tends to predict class 1 more frequently, which may happen if the dataset is imbalanced or if class 1 dominates the training data. Overall, the model shows strong performance on class 1 but struggles to correctly recognize class 0.

## MODEL EVALUATION

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_test_pred_knn)
```

```python
confusion_matrix(y_test, y_test_pred_dt)
```

```python
print(classification_report(y_test, y_test_pred_knn))
```

```python
print(classification_report(y_test, y_test_pred_dt))
```

The classification report shows that the model performs very well on class 1 but poorly on class 0. Precision, recall, and F1-score for class 1 are high, meaning the model is good at correctly identifying positive cases. However, the scores for class 0 are very low, which indicates the model struggles to detect this class. This happens because the dataset is imbalanced (class 1 has much more data than class 0), causing the model to learn more about the majority class. The weighted average adjusts the metrics based on the number of samples in each class, giving a more realistic overview of the model's overall performance, while macro average treats both classes equally. Overall, although the accuracy is 71%, the model is biased toward predicting class 1.

```python
from sklearn.model_selection import GridSearchCV
# GridSearchCV searches all parameter combinations to find the best model settings.
```

```python
knn_params = {
    "n_neighbors": [3, 7, 9, 10],
    "p": [2, 4, 6]
}
```

```python
knn_new = GridSearchCV (KNeighborsClassifier(), knn_params)
```

```python
knn_new.fit(X_train_scaled, y_train)
```

```python
# Displays the best hyperparameter combination found during the grid search
knn_new.best_params_
```

```python
knn_new_best = knn_new.best_estimator_
```

```python
knn_new_best.predict(X_test_scaled)
```

```python
probabilty = knn_new_best.predict_proba(X_test_scaled)[:, 1]
```

```python
(probabilty >= 0.5).astype("int")
```

```python
knn_new_best.predict(X_test_scaled)
```

```python
recalls = []
precisions = []
accuracies = []
f1_scores = []

for i in range(1, 11):
    y_best_knn_proba = knn_new_best.predict_proba(X_test_scaled)[:, 1]

    recalls.append(recall_score(y_test, y_best_knn_proba >= i/10))
    precisions.append(precision_score(y_test, y_best_knn_proba >= i/10))
    accuracies.append(accuracy_score(y_test, y_best_knn_proba >= i/10))
    f1_scores.append(f1_score(y_test, y_best_knn_proba >= i/10))
```

```python
metrics = pd.DataFrame ({
    "recall": recalls,
    "precition": precisions,
    "accuracy": accuracies,
    "f1_score": f1_scores
})
```
